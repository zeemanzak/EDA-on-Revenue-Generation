```python
import pandas as pd
```

```python
product_info=[{"productId":23, "name":"computer", "wholesale_price":500, "retail_price":1000, "sales":100},
              {"productId":96, "name":"python workout", "wholesale_price":35, "retail_price":75, "sales":1000},
              {"productId":97, "name":"pandas workout", "wholesale_price":35, "retail_price":75, "sales":500},
              {"productId":15, "name":"bananna", "wholesale_price":0.5, "retail_price":1, "sales":200},
              {"productId":87, "name":"sandwich", "wholesale_price":3, "retail_price":5, "sales":300}]
product_info
```

```python
df=pd.DataFrame(product_info)
```

```python
#total profit for each products
df["net_revenue_per_product"]=(df["retail_price"]-df["wholesale_price"]).mul(df["sales"])
#df["wholesale_price"]
total_profit_for_each_product=df[["net_revenue_per_product"]]
total_profit_for_each_product
```

```python
#total revenue for all the product
total_revenue_for_all_products=df["net_revenue_per_product"].sum()
total_revenue_for_all_products
```

```python
#product with retail price more than twice the wholesales price
product=(df["retail_price"]).gt(df["wholesale_price"].mul(2))
df[product]
```

```python
#amount made from food,book and computer
total_amount_by_category=df.groupby("name")["net_revenue_per_product"].sum().sort_values()
print(total_amount_by_category)
```

```python
#new net revenue after 30% on wholesales price
discount=0.3
df["discount_price"]=df["wholesale_price"].mul(1-0.3)
new_wholesales_price=df["discount_price"]
df["new_net_revenue"]=(df["retail_price"].sub(new_wholesales_price))*df["sales"]
total_net_revenue=df["new_net_revenue"].sum()
total_net_revenue


```

```python
#net revenue after 15% sales tax
tax_rate=0.15
df["sales_tax"]=df["new_net_revenue"]*tax_rate
df["sales_tax"]
df["current_income"]=df["new_net_revenue"]-df["sales_tax"]
df["current_income"] #current income after tax deduction of 15%
df
```
